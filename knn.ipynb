```python
mushroom_input = df.drop(['class'], axis = 1)
mushroom_target = df['class']
```

```python
import pandas as pd
mushroom_input = pd.get_dummies(mushroom_input)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
mushroom_target = encoder.fit_transform(mushroom_target)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df_encoded = df.apply(label_encoder.fit_transform)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

train_input, test_input, train_target, test_target = train_test_split(mushroom_input, mushroom_target, test_size = 0.3)
train_input.shape, test_input.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()
kn.fit(train_input, train_target)
```

```python
y_pred = kn.predict(test_input)
print("prediction accuracy: {:.2f}".format(np.mean(y_pred == test_target)))
y_pred = kn.predict(train_input)
print("prediction accuracy: {:.2f}".format(np.mean(y_pred == train_target)))
```

```python
import matplotlib.pyplot as plt

x_feature = train_input['odor_n']
y_feature = train_input['gill-size_b']


target = train_target

plt.figure(figsize=(8, 6))
scatter = plt.scatter(x_feature, y_feature, c=target, cmap='coolwarm', alpha=0.7)
plt.colorbar(scatter, label="Class (0: Edible, 1: Poisonous)")
plt.title("Scatter Plot of Odor_n vs. Gill-Size_b")
plt.xlabel("Odor_n (Encoded Feature)")
plt.ylabel("Gill-Size_b (Encoded Feature)")
plt.grid(True)
plt.show()
```
